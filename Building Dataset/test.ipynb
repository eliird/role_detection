```python
def cart2sph(x,y,z):
    XsqPlusYsq = x**2 + y**2
    r = math.sqrt(XsqPlusYsq + z**2)               # r
    elev = math.atan2(z,math.sqrt(XsqPlusYsq))     # theta
    az = math.atan2(y,x)                           # phi
    return (elev, az)

def cart2sphPose(pose:np.array):
    cart_pose = []
    for person in pose:
        p = []
        for frame in person:
            j = []
            for joint in frame:
                j.append(cart2sph(joint[0], joint[1], joint[2]))
            p.append(j)
        cart_pose.append(p)
    return np.array(cart_pose)
        
```

```python
import torch
import numpy as np
from torch.utils.data import Dataset
import pickle
import pandas as pd

def loadPickle(path):
    fileHandle = open(path, 'rb')
    data = pickle.load(fileHandle)
    fileHandle.close()
    return data

class custom_data(Dataset):
    def __init__(self, path):
        self.data = loadPickle(path)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        gaze_array = torch.flatten(torch.Tensor(self.data['Gaze Data'][idx]))
        pose_array = torch.DoubleTensor(self.data['Pose Data'][idx])
        label = torch.tensor(self.data['Encoded Future Roles'][idx])

        return (gaze_array, pose_array, label)
```

```python
class VGG16(nn.Module):
    def __init__(self, num_classes=10):
        super(VGG16, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(3, 8, kernel_size=3, stride=1, padding=0),
            nn.BatchNorm2d(8),
            nn.ReLU())
        self.layer2 = nn.Sequential(
            nn.Conv2d(8, 4, kernel_size=3, stride=1, padding=0),
            nn.BatchNorm2d(4),
            nn.ReLU(), 
            nn.MaxPool2d(kernel_size = 2, stride = 2))
        self.layer3 = nn.Sequential(
            nn.Conv2d(4, 4, kernel_size=3, stride=1, padding=0),
            nn.BatchNorm2d(4),
            nn.ReLU())
        self.layer4 = nn.Sequential(
            nn.Conv2d(4, 4, kernel_size=3, stride=1, padding=0),
            nn.BatchNorm2d(4),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2))
        self.fc = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(4*53*53, 512),
            nn.ReLU())
        self.fc1 = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(512, 256),
            nn.ReLU())
        self.fc2= nn.Sequential(
            nn.Linear(256, num_classes))
    def forward(self, x):
        out = self.layer1(x)
        #print("Layer 1 ",out.shape)
        out = self.layer2(out)
        #print("Layer 2 ",out.shape)
        out = self.layer3(out)
        #print("Layer 3 ",out.shape)
        out = self.layer4(out)
        #print("Layer 4 ",out.shape)
        out = out.reshape(out.size(0), -1)
        #print("Layer 5 ",out.shape)
        out = self.fc(out)
        out = self.fc1(out)
        out = self.fc2(out)
        return out
```

```python
from flash_pytorch import FLASH

class FTransformer(torch.nn.Module):
    def __init__(self, input_dim, num_classes):
        super(FTransformer, self).__init__()
        self.flash1 = FLASH(
            dim = 6,
            group_size = 256,             # group size
            causal = True,                # autoregressive or not
            query_key_dim = 128,          # query / key dimension
            expansion_factor = 2.,        # hidden dimension = dim * expansion_factor
            laplace_attn_fn = True   # new Mega paper claims this is more stable than relu squared as attention function
                    )
        
        self.flash2 = FLASH(
            dim = 6,
            group_size = 256,             # group size
            causal = True,                # autoregressive or not
            query_key_dim = 128,          # query / key dimension
            expansion_factor = 2.,        # hidden dimension = dim * expansion_factor
            laplace_attn_fn = True   # new Mega paper claims this is more stable than relu squared as attention function
                    )
        self.flat = nn.Flatten()  
        self.fc_layer = nn.Sequential(
                            nn.Linear(12240, 4000),
                            nn.BatchNorm1d(4000),
                            nn.ReLU(),
                            nn.Linear(4000, 1024),
                            nn.BatchNorm1d(1024),
                            nn.ReLU(),
                            nn.Linear(1024, 512),
                            nn.BatchNorm1d(512),
                            nn.ReLU(),
                            nn.Linear(512, 256),
                            nn.BatchNorm1d(256),
                            nn.ReLU(),
                            nn.Linear(256,128),
                            nn.BatchNorm1d(128),
                            nn.ReLU(),
                            nn.Linear(128,64),
                            nn.BatchNorm1d(64),
                            nn.ReLU(),
                            nn.Linear(64,32),
                            nn.BatchNorm1d(32),
                            nn.ReLU(),
                            nn.Linear(32,num_classes),
                            nn.ReLU()
            # Adjust 28 based on the chosen kernel size and sequence length
                                    ) 
    def forward(self, x1, x2):
        x1 = self.flash1(x1)
        x2 = self.flash2(x2)
        x =  self.flat(torch.cat([x1, x2], dim = 1))
        #print(x.shape)
        x = self.fc_layer(x)
        return x
```
